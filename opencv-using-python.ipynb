# Define a  user defined function, catimageShow() and convert BGR image to RGB image 

```python
def catimageShow(imageTitle,image):
    imageVariable = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(4,2))
    plt.imshow(imageVariable)
    plt.title(imageTitle)
    plt.show()
```

# Rotation of image  based on rotation  angle .Using warpAffine() function 

```python
(imageH,ImageW) = imagedata.shape[:2]
```

```python
centerX,centerY = (imageH/2,ImageW/2)
```

```python
imageRotate = cv2.getRotationMatrix2D((centerX,centerY),90,1.0)
rotateNow = cv2.warpAffine(imagedata,imageRotate,(ImageW,imageH))
catimageShow( " Rotate image",rotateNow)
```

```python
imageRotate = cv2.getRotationMatrix2D((centerX,centerY),180,1.0)
rotateNow = cv2.warpAffine(imagedata,imageRotate,(ImageW,imageH))
catimageShow( " Rotate image",rotateNow)
```

```python
imageRotate = cv2.getRotationMatrix2D((centerX,centerY),270,1.0)
rotateNow = cv2.warpAffine(imagedata,imageRotate,(ImageW,imageH))
catimageShow( " Rotate image",rotateNow)
```
